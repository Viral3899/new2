```python
#Main title
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([2,5,3,8])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Title 1")

#plot 2:
x = np.array([0, 2,5,3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("Tile 2")

plt.suptitle("Main Title")
plt.show()
```

```python
y = np.array([20,35,10,45])
mylabels = ["USA", "India", "Russia", "China"]

plt.pie(y, labels=mylabels,autopct='%1.2f%%')  # autopct for show percentage
plt.show() 
```

```python
plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

```python

myexplode = [0, 0.1, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

```python
mycolors = ["black", "pink", "b", "yellow"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

```python
plt.pie(y, labels = mylabels)
plt.legend(title="Country",loc="upper center",bbox_to_anchor=(1.1, 1.15),ncol=4)
plt.show()
```
