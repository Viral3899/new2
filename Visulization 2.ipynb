```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips[(tips["day"]=="sun")&(tips["smoker"]=="Yes")]
```

```python

sns.relplot(x="total_bill", y="tip", hue="day", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="day",style="smoker", data=tips)
plt.figure(figsize=(15,10))
```

```python
sns.relplot(data=tips,x="total_bill",y='tip',kind='line')
```

```python
sns.jointplot(x='tip',y='size',data=tips,kind='hex') #try with total bill
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex',color="pink") #try with total bill
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='reg',color="green")
plt.figure(figsize=(15,8))
```

```python
plt.figure(figsize=(15,8))
sns.pairplot(tips,hue="sex")
```

```python
plt.figure(figsize=(15,8))
sns.pairplot(tips,kind="reg")
```

```python
plt.figure(figsize=(15,8))
sns.pairplot(tips,hue="smoker")
```

```python
plt.figure(figsize=(15,8))
sns.distplot(tips['total_bill'])
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(tips.corr(),annot=True)
```

```python
plt.figure(figsize=(15,8))
sns.countplot('smoker',data=tips)
```

```python
plt.figure(figsize=(15,8))
sns.countplot('day',data=tips)
```
