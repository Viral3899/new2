<h3> Function For Convert "?" Into "NAN"  And Replace "NAN" with Mean of Column</h3>

```python
def fillmissing(colname,df):
    
    print(df.dtypes)
    print()
    
    print(df[colname].unique())
    print()
    
    df[colname]=df[colname].replace("?",np.nan)
    print()
    
    print(df.isnull().sum())
    print()
    
    df[colname]=df[colname].astype(float)
    print()
    
    print(df.dtypes)
    print()
    
    df[colname]=df[colname].fillna(df[colname].mean())
    print()
    
    print(df[colname].unique())
    print()
    
    return df
```
