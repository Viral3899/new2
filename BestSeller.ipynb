Setting up Parameter for Visulization

```python
plt.rcParams["figure.figsize"]=(20,15)
plt.rcParams["font.size"]=(10)
plt.rcParams["xtick.labelsize"]=(15)
plt.rcParams["ytick.labelsize"]=(15)
plt.rcParams["font.weight"]=("normal")
plt.rcParams['xtick.alignment']="center"
plt.rcParams['ytick.alignment']="center"
plt.rcParams['font.family']="Tahoma"
plt.rcParams['grid.color']="black"
plt.rcParams["grid.alpha"]=True
plt.rcParams["legend.fontsize"]=10
plt.rcParams["legend.title_fontsize"]=10
```

<h2>Histograms or Densityplot</h2>

#### Histogram of Price

```python
sns.histplot(df["Price"],color="yellowgreen")
plt.grid()
plt.xlabel("Price",fontsize=30)
plt.ylabel("Count",fontsize=30)
```

Most Book Price range Lie between <b>5 & 15</b>

#### Histogram of User Rating

```python
sns.histplot(data=df,x="User Rating",bins=10,color="aquamarine")
```

Most of books Rated Between <b>4.6 to 4.8</b>

#### Desityplot of Review

```python
sns.distplot(x=df["Reviews"],color="orchid")
```

Graph Shows Numberof <b>Reviews</b> for Book <b>Mostly</b> Ranges <b>5000 to 10000</b>

#### Histograms For All Columns

```python
df.hist(figsize=(20,10),color="lightcoral");
```

```python
df["User Rating"].value_counts().plot(kind="bar",color="orange",figsize=(20,15))
plt.legend(fontsize=15)
plt.xlabel("User Rating",fontsize=35)
plt.ylabel("Count",fontsize=35)
plt.xticks(fontsize=20,rotation=0)
plt.yticks(fontsize=20)
plt.grid(color="skyblue");
```

## Count Plot For Genre

```python
sns.countplot(data=df,x="Genre",palette="summer")
plt.legend(fontsize=15)
plt.xlabel("Genre",fontsize=20)
plt.ylabel("Frequency",fontsize=20)
plt.xticks(fontsize=20,rotation=0)
plt.yticks(fontsize=20);
```

#### Countplot for User Rating

```python
sns.countplot(data=df,x="User Rating",palette="inferno")
plt.legend(fontsize=15)
plt.xlabel("Genre",fontsize=20)
plt.ylabel("Frequency",fontsize=20)
plt.xticks(fontsize=20,rotation=0)
plt.yticks(fontsize=20)
plt.grid();
```

▶ More than <b>120 books have rating of 4.8</b> <br>
▶ Around <b>50 Books Have rating of 4.9</b>

```python
sns.heatmap(df.corr(),cmap="vlag")
plt.xticks(fontsize=20,rotation=0)
plt.yticks(fontsize=20);
```

There is <b>No such Correlation</b> in Data

```python
df.groupby(["Genre"])["Reviews"].sum().plot(kind="pie",cmap="inferno",wedgeprops=dict(width=.15),autopct="%3.1f%%",figsize=(10,8),shadow=True)
plt.grid(color="darkred")
plt.legend(loc="upper center",bbox_to_anchor=(0.5, 1.15), ncol=7)
plt.xlabel("Genre",fontsize=20)
```

```python
df.groupby(["User Rating"])["Reviews"].sum().plot(kind="bar",color="gold",figsize=(10,8))
plt.grid(color="silver")
plt.legend(loc="upper center",bbox_to_anchor=(0.5, 1.15), ncol=7)
plt.xlabel("User rating",fontsize=20)
plt.ylabel("Reviews",fontsize=20)
```

```python
df.groupby(["Genre"])["User Rating"].mean().plot(kind='bar',color="violet")
plt.ylabel("User rating",fontsize=30)
plt.xlabel("Genre",fontsize=30)
plt.xticks(fontsize=30,rotation=0)
plt.yticks(fontsize=30)
plt.grid();
print(df.groupby(["Genre"])["User Rating"].mean())
print("Average Rating Of Fiction And Non Fiction Books")
```
